# groupby()

```python
#для сбора групп и строк вместе и вызова для них агрегатной функции


import numpy as np
import pandas as pd
```

1.Создать словарь

```python
dict = {'Company':['COMP1','COMP1','COMP2','COMP2','COMP3','COMP3'],
        'Months': ['JAN','FEB','MAR','APR','MAY','JUN'], 
        'Salary': [2000, 1500, 3000, 5000, 2500, 3500]}
dict
```

2.Из словаря сформировать DataFrame

```python
df = pd.DataFrame(dict)
df
```

3.Получить  значение для каждой компании


```python
#среднее
df.groupby('Company').mean()
```

```python
#сумму
df.groupby('Company').sum()
```

```python
#среднее квадратичное отклонение
df.groupby('Company').std()
```

```python
#точечноя выборка
df.groupby('Company').sum().loc['COMP3']
```

```python
#подсчет объектов
df.groupby('Company').count()
```

```python
#максимальное и минимальное значение
df.groupby('Company').max()
```

```python
df.groupby('Company').min()
```

описать подробно

```python
df.groupby('Company').describe()
```

# merge(слияние)

```python
df1 = pd.DataFrame({'order_id': [114, 235, 432],
                     'user_id': [1, 2, 3],
                     'user_name': ['James Brown', 'Jack White',
                                   'Jane Green'],
                     'country': ['USA', 'USA', 
                                    'France']})

df2 = pd.DataFrame({'order_id': [114, 235, 432],
                     'user_id': [1, 2, 3],
                     'order_date': ['2020-02-11', '2020-02-11', 
                                    '2020-02-15'],
                     'OS': ['Android', 'iOS', 'Android']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='user_id')
```

```python
pd.merge(df1,df2,on='order_id')
```

# join()

```python
df1 = pd.DataFrame({
                     'user_name': ['James Brown', 'Jack White',
                                   'Jane Green'],
                     'country': ['USA', 'USA', 
                                    'France']},
                  index=['ind1', 'ind2', 'ind3'])

df2 = pd.DataFrame({'order_id': [114, 235, 432],
                     'user_id': [2, 4, 3],
                     'order_date': ['2020-02-11', '2020-02-11', 
                                    '2020-02-15'],
                     'OS': ['Android', 'iOS', 'Android']},
                    index=['ind1', 'ind4', 'ind3'])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```
