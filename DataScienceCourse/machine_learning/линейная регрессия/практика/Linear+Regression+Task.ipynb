### 16. Now let's split the data into training and testing sets. Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column.
Теперь давайте разделим данные на обучающие и тестовые наборы. Установите переменную X, равную числовым характеристикам клиентов, и переменную y, равную столбцу "Yearly Amount Spent"("Потраченная сумма за год").

```python
EC.columns
```

```python
#Вырезать не числовые значения'Email', 'Address', 'Avatar',
X = EC[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
y = EC['Yearly Amount Spent']
```

```python
EC.head()
```

### 17. Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set test_size=0.4 and random_state=101
Используйте model_selection. Используйте model_selection.train_test_split из sklearn для разделения данных на обучающие и тестовые наборы. Установите test_size=0,4 и random_state=101

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
X_train, X_test, y_train, y_test
```

### 18. Now its time to train the model on training data. Import LinearRegression from sklearn.linear_model 
Теперь пришло время обучить модель на обучающих данных. Импортируйте линейную регрессию из sklearn.linear_model

```python
from sklearn.linear_model import LinearRegression
```

### 19. Create an instance of a LinearRegression() model and assign to a variable
Создайте экземпляр модели LinearRegression() и назначьте переменной

```python
lrm = LinearRegression()
```

### 20. Train/fit the model on the training data
Обучать/подгонять модель по данным обучения

```python
lrm.fit(X_train, y_train)
```

### 21. Print out the coefficients of the model
Распечатайте коэффициенты модели

```python
print('Coefficients: ', lrm.coef_)
```

### 22. Now that we have fit our model, let's evaluate its performance by predicting off the test values. Use predict() to predict off the X_test set of the data
Теперь, когда мы подобрали нашу модель, давайте оценим ее производительность, предсказав тестовые значения. Используйте функцию predict() для прогнозирования набора данных X_test¶

```python
predictions = lrm.predict(X_test)
predictions
```

### 23. Create a scatterplot of the real test values vs the predicted values
Создайте диаграмму рассеяния реальных значений теста по сравнению с прогнозируемыми значениями

```python
plt.scatter(y_test, predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

### 24. Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error
Вычислите Среднюю Абсолютную ошибку, Среднеквадратичную ошибку и Среднеквадратичную ошибку

```python
from sklearn import metrics
#Абсолютная средняя ошибк
print('MAE: ', metrics.mean_absolute_error(y_test, predictions))
#средняя квадратичная ошиб
print('MSE: ', metrics.mean_squared_error(y_test, predictions))
#квадратный корень из средней квадратичной ошибки
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

### 25. Let's quickly explore the residuals to make sure everything was okay with our data. Plot a histogram of the residuals and make sure it looks normally distributed. Use either seaborn distplot, or just plt.hist()
Давайте быстро изучим остатки, чтобы убедиться, что с нашими данными все в порядке. Постройте гистограмму остатков и убедитесь, что она выглядит нормально распределенной. Используйте либо карту распределения seaborn, либо просто plt.hist()¶

```python
sns.histplot((y_test - predictions), bins=50)
```

### 26. Let's see how we can interpret the coefficients. Recreate the dataframe below
Давайте посмотрим, как мы можем интерпретировать коэффициенты. Воссоздайте фрейм данных ниже¶

```python
#Создать датафрейм с передачей кофициентов в названную и указаннцю колонку
cdf = pd.DataFrame(lrm.coef_, X.columns, columns=['Coefficients'])
#cdf = ['Coefficients']
cdf
```
