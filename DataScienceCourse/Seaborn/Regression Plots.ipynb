```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], 
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time', hue="sex")
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', hue="day")
```

```python
# aspect- 
#height-
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue="sex", aspect=0.3, height=6)
```
