```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
import numpy as np
#Среднее значение по категории(ям)
# estimator-это оценщик ,сюда можно передовать собственную функцию изменение defolt функции(mean-среднее значение)
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
#Среднее значение по категории(ям)
# estimator-это оценщик который явно считает количество значений сколько раз оно встречается
#тоже как pandas.count()

sns.countplot(x='sex', data=tips)
```

```python
#Для распределения категорийных данных
#Показывают ядро, усы и выбросы(исключительные ситуации)
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
#dodge==split
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
#комбинирование методов
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
#обобщающая функция
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
#Более новая версия 
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```
