```python
#импортируем библиотеку
import seaborn as sns
```

```python
#Для визуализации графиков внутри jupyter 
%matplotlib inline
```

```python
#Получить и загрузить встроенные данные для дальнейшей их обработки
#seaborn default datasets
#seaborn.load_dataset(name(str), cashe(bool)=True, data_home(str)=None,**kws(dict) )
dots = sns.load_dataset('dots')
```

```python
#Определить первые пять строк данных таблицы
dots.head()
```

```python
#распределить одну переменную из таблицы где:
#sns-библиотека Seaborn
#distplot-распределение
#dots-имя таблицы
#['firing_rate']-столбец таблицы
#kde-распределение выборки(точность),логический тип указывающий нужно ли рисовать кривую оценки плотности ядра Гауса
#bins-бункеры.Используются для управления количества столбцов.Чем выше значение ,тем больше детализация
#hist-логический тип указывающий нужно ли рисовать(маркировать) гистограмму

#sns.distplot(dots['firing_rate'])
sns.distplot(dots['firing_rate'], kde=False, bins=20)
```

```python
#jointplot-позволяет комбинировать(объединять) два разных графика распределения
#х и у - это элементы сравнения
#data= - таблица содержащая эти элементы

sns.jointplot(x='coherence', y='firing_rate', data=dots)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=20)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
# парный график
sns.pairplot(tips)
```

```python
# парный график
#hue-категория имя столбца

sns.pairplot(tips, hue='sex')
```

```python
# парный график где:
#tips-имя таблицы DataFrame
#hue-категория имя столбца
#palette-цветовая палитра

sns.pairplot(tips, hue='sex', palette='husl')
```
