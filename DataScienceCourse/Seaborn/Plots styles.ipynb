```python
#импортировать библиотеку
import seaborn as sns
#команда для визуализации графиков
%matplotlib inline
#скачать и установить из библиотеки seaborn Dataset 'tips'
tips = sns.load_dataset('tips')
#просмотреть верхние пять строк базы
tips.head()
```

```python
#Стиль
sns.set_style('white')
#shift+tab  style : dict, None, or one of {darkgrid, whitegrid, dark, white, ticks}
#График
sns.countplot(x='sex', data=tips)
#Убрать рамку сверху и слева по умалчанию
sns.despine(left=True, bottom=True)
```

```python
#Применение matplotlib для увеличения картинки и графика
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))
sns.countplot(x='sex', data=tips)
```

```python
#font_scale-размер шрифта
#context : dict, None, or one of {paper, notebook, talk, poster}
sns.set_context('notebook', font_scale=1)
sns.countplot(x='sex', data=tips)
```

```python
#Цвета
#matplotlib.org/colormaps
# hue-разделение по полу
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='cool')
```
